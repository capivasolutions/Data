# Notebook onde analizaremos quais são os atributos relevantes

```python
import numpy as np
import pandas as pd # importando a lib pandas para armazenar os dados
import matplotlib.pyplot as plt # importando a bib matplotlib para visualizar os dados em gráficos
import seaborn as sb # important a lib seaborn para visualização de gráficos

df = pd.read_csv('../Bases/creditcard.csv')

df.head()

data = df.to_numpy()

legitimos = data[data[:,30] == 0]
fraudes = data[data[:,30] == 1]

print(len(legitimos))
print(len(fraudes))

colunas = {"Time": 0,
           "A1": 1,
           "A2": 2,
           "A3": 3,
           "A4": 4,
           "A5": 5,
           "A6": 6,
           "A7": 7,
           "A8": 8,
           "A9": 9,
           "A10": 10,
           "A11": 11,
           "A12": 12,
           "A13": 13,
           "A14": 14,
           "A15": 15,
           "A16": 16,
           "A17": 17,
           "A18": 18,
           "A19": 19,
           "A20": 20,
           "A21": 21,
           "A22": 22,
           "A23": 23,
           "A24": 24,
           "A25": 25,
           "A26": 26,
           "A27": 27,
           "A28": 28,
           "Amount": 29,
           "Classe": 30}
```

## Abaixo plotaremos as colunas comparando as com a classe

### Time

```python
# Plotando a coluna "Time" com a "Class"

plt.plot(legitimos[:, colunas['Time']], legitimos[:, colunas['Classe']], 'ob', label="Legitimos")
plt.plot(fraudes[:, colunas['Time']], fraudes[:, colunas['Classe']], 'or', label="Fraudes")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()

violinos = plt.violinplot([legitimos[:, colunas['Time']], fraudes[:, colunas['Time']]], showmeans=True);
cores = ['Blue', 'Red']

for pc, cor in zip(violinos['bodies'], cores):
    pc.set_facecolor(cor)

plt.legend(['Legitimos', 'Fraudes'], loc='center')
plt.show()
```

### Parallel line plot

```python
# Plotando a coluna "A20" com a "Class"

plt.figure(figsize=(10,10))

cFraude = fraudes[:, 1:]
cFraude = cFraude[:, :-2]

cLegitimo = legitimos[:, 1:]
cLegitimo = cLegitimo[:, :-2]


for i in cFraude:
    plt.plot(i, '-r', label="Fraudes")

for i in cLegitimo:
    plt.plot(i, '-b', label="Legitimos")

plt.show()
```
